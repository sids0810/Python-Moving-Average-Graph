<h4>Step 1: import necessary libraries</h4>

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
%matplotlib inline
```

<h4>Step 2: Ask the user for inputs: stock ticker, start date, end date</h4>

```python
input_stock = input("Enter the stock ticker: ")
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")
```

<h4>Step 3: Use datetime to convert the start_date to a datetime object. To calculate moving averages accurately, we need data from the period preceding the start date as well. Hence, we use the datetime methods to change the start date.</h4>

```python
format_date = "%Y-%m-%d"
#Create datetime object
start_date = datetime.datetime.strptime(start_date,format_date) 
#Define the period you want to go back
preceding_period = datetime.timedelta(days = 365)
#Move the start date back
start_date_new = start_date - preceding_period
#Convert the datetime object back to string
start_date_new = start_date_new.strftime(format_date)
```

<h4>Step 4: Create a yfinance object and import historical stock data using the method .history()</h4>

```python
#Create the yfinance object
stock_object = yf.Ticker(input_stock)
#Import historical data using .history()
stock_historical = stock_object.history(start = start_date_new , end = end_date, interval ="1d")
#Display the imported data to confirm
stock_historical
```

<h4>Step 5: Calculate the Moving Averages (50 and 200 sma)</h4>

```python
#Use the .rolling() method to calculate the mean of n preceding values. This is how we calculate the moving averages.
#This also creates new columns in the DataFrame to store the MAs.
stock_historical['50SMA'] = stock_historical['Close'].rolling(window=50).mean()
stock_historical['200SMA'] = stock_historical['Close'].rolling(window=200).mean()
stock_historical
```

<h4>In the step performed above, you see the NaN - Not a Number outputs for the first few rows. This is because the rolling period is set to 50 or 200. Due to insufficient preceding date for those specific start dates, pandas is unable to calculate a moving average. This is exactly why we had to make the start date adjustments before.</h4>

<h4>Step 6: Remove all the preceding data and make the start date first row in the data frame. Visualize Closing Price, 50SMA, and 200SMA</h4>

```python
#Create a subset of the DataFrame to remove all the preceding data which was used only for calculations. 
stock_historical_new = stock_historical.loc[pd.to_datetime(start_date, utc=True):]
#Use matplotlib to create a line graph. Label the graph and its axes appropriately. 
plt.plot(stock_historical_new[["Close","50SMA","200SMA"]])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.title("Chart with 50- and 200-SMA for "+input_stock)
plt.show()
```
